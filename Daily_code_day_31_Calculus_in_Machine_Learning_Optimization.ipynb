# Calculus in Machine Learning - Optimization

## 1. Optimizing a single variable function using Gradient Descent

```python
# Import libraries
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define f(x) and f'(x) function
def f(x):
  return x**4-3*x**3+2

def gradient(x):
  return 4*x**3-9*x**2

# Define Gradient Descent function
def gradient_descent(gradient, x_init, learning_rate=0.01, epochs=100):
  x = x_init
  history = [x]

  for _ in range(epochs):
    x = x-learning_rate*gradient(x)
    history.append(x)

  return x, history
```

```python
# Result
x_init = 0.5
learning_rate = 0.01
epochs = 100
x_optimal, history = gradient_descent(gradient, x_init, learning_rate, epochs)

# Plotting graph
x_vals = np.linspace(-1, 3, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x)")
plt.scatter(history, [f(x) for x in history], color="red", s=10, label="Iterations")
plt.scatter(x_optimal, f(x_optimal), color="green", marker="x", s=100, label=f"Min at x={x_optimal:.3f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

## 2. Linear Regression with Gradient Descent

```python
# Sample data
X = np.array([i for i in range(1, 6)])
y = np.array([i for i in range(2, 7)])
print(X, y)
```

```python
w, b = 0, 0
lr = 0.01
epochs = 100
```

```python
# Gradient Descent
m = len(X)
for _ in range(epochs):
    y_pred = w * X + b
    dw = (-2/m) * np.sum(X * (y - y_pred))
    db = (-2/m) * np.sum(y - y_pred)

    w -= learning_rate * dw
    b -= learning_rate * db

# Dự đoán giá trị mới
y_pred = w * X + b
```

```python
# Plotting graph
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

print(f"Hệ số tối ưu: w = {w:.3f}, b = {b:.3f}")
```
