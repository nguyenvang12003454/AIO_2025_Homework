# Calculus in Machine Learning - Basic Chain Rule

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Finite difference derivative function
def numerical_derivative(f, x, h=1e-5):
    return (f(x+h)-f(x))/(h)

# Graphing function
def plot_chain_rule(f, g, x_range=(-2, 2), title="Chain Rule derivative"):
  x_vals = np.linspace(x_range[0], x_range[1], 400)
  # Tính giá trị của g(x) và f(g(x))
  g_vals = g(x_vals)
  f_vals = f(g_vals)

  # Tính đạo hàm theo Chain Rule
  g_derivative_vals = numerical_derivative(g, x_vals)
  f_derivative_vals = numerical_derivative(f, g_vals)
  chain_rule_derivative_vals = f_derivative_vals * g_derivative_vals

  # Vẽ đồ thị của f(g(x)) và đạo hàm
  plt.figure(figsize=(8, 5))
  plt.plot(x_vals, f_vals, label="f(g(x))", color='blue')
  plt.plot(x_vals, chain_rule_derivative_vals, label="f'(x) theo Chain Rule", color='red', linestyle="--")

  plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
  plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

  plt.xlabel("x")
  plt.ylabel("Giá trị")
  plt.title(title)
  plt.legend()
  plt.grid()
  plt.show()
```

```python
# Vẽ đồ thị đạo hàm của hàm hợp
plot_chain_rule(lambda x: np.exp(x), lambda x: x**3-2*x+1, (-2, 2), title="f(x) = e^(x^3-2x+1)")
plot_chain_rule(lambda x: np.log(x), lambda x: 5*x**2+3*x+1, (-2, 2), title="f(x) = ln(5x^2+3x+1)")
plot_chain_rule(lambda x: np.sqrt(x), lambda x: x**4+2*x**2+1, (-2, 2), title = "f(x) = sqrt(x^4+2x^2+1)")
plot_chain_rule(lambda x: np.sin(x), lambda x: x**2-3*x+2, (-2, 2), title = "f(x) = sin(x^2-3x+2)")
plot_chain_rule(lambda x: np.exp(x), lambda x: np.log(x**2+4), (-2, 2), title = "f(x) = e^(ln(x^2+4))")
```

```python
plot_chain_rule(lambda x: 1/np.sqrt(x), lambda x: x**3+x+2, (-2, 2), title="f(x) = 1/sqrt(x^3+x+2)")
```
