# Calculus in Machine Learning

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Finite difference derivative function
def derivative(f, x, h=1e-5):
    return (f(x+h)-f(x))/(h)

# Graphing function
def plot_func_and_derivative(f, x_range=(-2, 2), title="f(x) and f'(x) graph"):
  x_vals = np.linspace(x_range[0], x_range[1], 400)
  y_vals = f(x_vals)
  y_derivative_vals = derivative(f, x_vals)

  plt.figure(figsize=(8, 5))
  plt.plot(x_vals, y_vals, label="f(x)", color='blue')
  plt.plot(x_vals, y_derivative_vals, label="f'(x) (numerical)", color='red', linestyle="--")
  plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
  plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
  plt.xlabel("x")
  plt.ylabel("f(x) & f'(x)")
  plt.title(title)
  plt.legend()
  plt.grid()
  plt.savefig(f'{title}.png')
  plt.show()
```

```python
# 1. Derivative of quadratic function
def quadratic(x):
  return x**2-3*x+2

plot_func_and_derivative(quadratic, (-5, 5), "Bài 1: f(x) = x^2 - 3x + 2")
```

```python
# 2. Derivative of cubic function
def cubic(x):
  return x**3-3*x**2+4*x-2

plot_func_and_derivative(cubic, (-2, 4), "Bài 2: f(x) = x^3 - 3x^2 + 4x - 2")
```

```python
# 3. Derivative of exponential function
def exponential(x):
  return np.exp(x)

plot_func_and_derivative(exponential, (-2, 4), "Bài 3: f(x) = e^x")
```

```python
# 4. Derivative of logarit function
def logarit(x):
  return np.log(x)

plot_func_and_derivative(logarit, (0.1, 3), "Bài 4: f(x) = ln(x)")
```

```python
# 5. Derivative of sin(x) function
def sin(x):
  return np.sin(x)

plot_func_and_derivative(sin, (-2*np.pi, 2*np.pi), "Bài 5: f(x) = sin(x)")
```

```python
# 6. Derivative of cos(x) function
def cos(x):
  return np.cos(x)

plot_func_and_derivative(cos, (-2*np.pi, 2*np.pi), "Bài 6: f(x) = cos(x)")
```

```python
# 7. Derivative of tan(x) function
def tan(x):
  return np.tan(x)

plot_func_and_derivative(tan, (-2*np.pi/5, 2*np.pi/5), "Bài 7: f(x) = tan(x)")
```

```python
# 8. Derivative of square root function
def square_root(x):
  return np.sqrt(x)

plot_func_and_derivative(square_root, (0.1, 3), "Bài 8: f(x) = sqrt(x)")
```
