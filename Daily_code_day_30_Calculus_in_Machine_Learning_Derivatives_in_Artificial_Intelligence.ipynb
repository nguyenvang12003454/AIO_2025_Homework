# Calculus in Machine Learning - Derivatives in Artificial Intelligence

## 1. Derivaitve of Sigmoid activation function

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid function
def sigmoid(x):
  return 1/(1+np.exp(-x))

# Derivative of Sigmoid function
def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
# Create sample data
x = np.array([-2,-1,0,1,2])
y = sigmoid(x)
dy_dx = sigmoid_der(x)

# lot sigmoid activation
plt.plot(x, y, label = "Sigmoid(x)")
plt.plot(x, dy_dx, label = 'Derivative of Sigmoid(x)')
plt.legend()
plt.show()
```

## 3. Derivative of ReLU activation function

```python
def relu(x):
  return np.maximum(0, x)

def relu_der(x):
  return np.where(x > 0, 1, 0)
```

```python
x = np.linspace(-8, 8, 100)
y = relu(x)
dy_dx = relu_der(x)

plt.plot(x, y, label = "ReLU(x)")
plt.plot(x, dy_dx, label = 'Derivative of ReLU(x)')
plt.legend()
```

## 4. Derivative of Softmax and Cross_Entropy

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x))

def softmax_der(x):
  return softmax(x)*(1-softmax(x))

x = np.linspace(-8, 8, 100)
y = softmax(x)
dy_dx = softmax_der(x)

plt.plot(x, y, label = "Softmax(x)")
plt.plot(x, dy_dx, label = 'Derivative of Softmax(x)')
plt.legend()
```

## 5. Derivative of Tanh activation function

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def tanh_der(x):
  return 1-tanh(x)**2

x = np.linspace(-8, 8, 100)
y = tanh(x)
dy_dx = tanh_der(x)

plt.plot(x, y, label = "Tanh(x)")
plt.plot(x, dy_dx, label = 'Derivative of Tanh(x)')
plt.legend()
```
